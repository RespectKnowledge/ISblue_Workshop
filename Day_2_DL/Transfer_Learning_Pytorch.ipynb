```python
import torch
import torch.nn as nn
import torchvision
dir(torchvision.models)
```

```python
model_conv = torchvision.models.resnet18(pretrained=True) # 
for param in model_conv.parameters():
    param.requires_grad = False  # freeze all weigts of the pre-trained models
```

```python
for name,layer in model_conv.named_children():
    print(name) # check the layers and their named
    #print(layer) # check layers
```

```python
num_ftrs = model_conv.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
model_conv.fc = nn.Linear(num_ftrs, 2) # 2 are the number of classes
print(model_conv)
```

```python
modelVGG=torchvision.models.vgg19(pretrained=True)

for name,layer in modelVGG.named_children():
    print(name)
    #print(layer)
    
numft=modelVGG.classifier[6].in_features
modelVGG.classifier[6]=nn.Linear(numft,2)
print(modelVGG)
```

```python
# fine-tuned model
modelDensNet=torchvision.models.densenet121(pretrained=True)
num_ftrs = modelDensNet.classifier.in_features
modelDensNet.classifier = nn.Linear(num_ftrs, 3) # number of classes=3
  
for name,layer in modelDensNet.named_children():
    print(name)
    #print(layer)  
```
